```python
import pandas as pd
import numpy as np
!pip install simpletransformers
from simpletransformers.classification import ClassificationModel,ClassificationArgs
from sklearn.model_selection import train_test_split
from sklearn.metrics import  accuracy_score
```

```python
data = pd.read_csv('https://github.com/clairett/pytorch-sentiment-classification/raw/master/data/SST2/train.tsv', delimiter='\t', header=None)
```

```python
data.columns = ['reviews','sentiment']
```

```python
data["sentiment"].value_counts()
```

```python
x=data["reviews"]
y=data["sentiment"]
```

```python
model_args=ClassificationArgs(overwrite_output_dir=True)
model_args.num_train_epochs=10
model_args.learning_rate=1e-5
```

```python
model=ClassificationModel("bert","bert-base-cased",num_labels=2,args=model_args,)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
train_df=pd.DataFrame({"text":x_train,"labels":y_train})
eval_df=pd.DataFrame({"text":x_test,"labels":y_test})
```

```python
train_df.head()
```

```python
model.train_model(train_df,acc=accuracy_score)
```

```python
result,model_output_prob,wrong_predictions=model.eval_model(eval_df,acc=accuracy_score)
```
