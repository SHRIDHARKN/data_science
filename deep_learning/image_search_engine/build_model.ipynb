```python
import pandas as pd
import numpy as np
import joblib
from sklearn.neighbors import NearestNeighbors
import warnings
warnings.simplefilter("ignore")
```

```python
resnet_features = joblib.load("data/features/resnet50_img_feature_extracts")
```

```python
df = pd.DataFrame(resnet_features).T
df.head()
```

```python
df.columns = [f"feature_{x}" for x in range(df.shape[1])]
```

```python
df.head()
```

```python
model = NearestNeighbors(n_neighbors=5).fit(df)
distances, indices = model.kneighbors(df)
```

```python
df_neigh = pd.DataFrame(distances)
df_neigh.columns = [f"neigh_{x}" for x in range(df_neigh.shape[1])]
df_neigh.head()
```

```python
df_img_name = pd.DataFrame(indices)
df_img_name.columns = [f"image_name_{x}" for x in range(df_img_name.shape[1])]
df_img_name.head()
```

```python
joblib.dump(model,"data/models/knn_model")
joblib.dump(df,"data/models/knn_data")
```
